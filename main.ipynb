### Challenge 2: Count Commits

Count how many commits were made in the past week.

1. Obtain all the commits made in the past week via API, which is a JSON array that contains multiple commit objects.

````python
url='https://api.github.com/repositories/153720804/commits'

response = requests.get(url)
result=response.json()

df_commits=pd.DataFrame()
df_commits=json_normalize(result)
df_commits.head()
````

````python
df_commits.columns
````

````python
print(df_commits.columns)
df_commits.columns=[str.strip(x.lower()) for x in df_commits.columns]
df_commits.columns=[re.sub(r"[. \ :+]","_",x) for x in df_commits.columns]
df_commits.columns=[re.sub("__","_",x) for x in df_commits.columns]
print(df_commits.columns)
````

### 2. Count how many commit objects are contained in the array.

````python
df_commits[['commit_author_date', 'commit_author_email', 'commit_author_name',
       'commit_comment_count', 'commit_committer_date',
       'commit_committer_email', 'commit_committer_name', 'commit_message',
       'commit_tree_sha', 'commit_tree_url', 'commit_url',
       'commit_verification_payload', 'commit_verification_reason',
       'commit_verification_signature', 'commit_verification_verified']].count()
````
